```python
import pyaudio
import wave
```

```python
def playSound(filePath): ##function used to play .wav files. Takes file directory path as a parameter
    #define stream chunk   
    chunk = 1024  
    #open a wav format music  
    f = wave.open(r"%s"%(filePath),"rb")
    #instantiate PyAudio  
    p = pyaudio.PyAudio()  
    #open stream  
    stream = p.open(format = p.get_format_from_width(f.getsampwidth()),  
                    channels = f.getnchannels(),  
                    rate = f.getframerate(),  
                    output = True) 
    #read data  
    data = f.readframes(chunk)  

    #play stream  
    while data:  
        stream.write(data)  
        data = f.readframes(chunk)  

    #stop stream  
    stream.stop_stream()  
    stream.close()  

    #close PyAudio  
    p.terminate()
    
def bic_bac_boe():
    gameEnd = False # bool initialized to keep the program running while the game is not over
    win_combinations = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    board = [1, 2, 3, 4, 5, 6, 7, 8, 9] # array initialized for positions on the board
   

    def draw(): #print each cell in the board
        print(board[0], board[1], board[2])
        print(board[3], board[4], board[5])
        print(board[6], board[7], board[8])
        print()
        
    def p1(): #function for player one
        n = ChooseNumber()
        if board[n] == "X" or board[n] == "O":
            print("\nYou can't go there")
            p1()
        else:
            board[n] = "X" #set board of index n equal to an X

    def p2(): #function for player two
        n = ChooseNumber()
        if board[n] == "X" or board[n] == "O":
            print("\nYou can't go there")
            p2()
        else:
            board[n] = "O" #set board of index n equal to a circle

    def ChooseNumber(): # function to choose a number
        while True:
            while True:
                a = input()
                try:
                    a  = int(a) #rejects strings
                    a -= 1 
                    if a in range(0, 9):
                        return a
                    else:
                        print("\nThat's not on the board")
                        continue
                except ValueError:
                   print("\nThat's not a number")
                   continue

    def check_board():
        count = 0 ##used to keep track of turns (max is 9)
        for a in win_combinations: ## 
            if board[a[0]] == board[a[1]] == board[a[2]] == "X": ##if there are three X's player one wins
                print("player 1 wins\n")
                print("yeet!\n")
                playSound("C:/Users/Fred Devereaux/Downloads/Microsoft Windows XP Startup Sound.wav")
                return True

            if board[a[0]] == board[a[1]] == board[a[2]] == "O": ##if there are three O's player two wins
                print("player 2 wins\n")
                print("yeet!\n")
                playSound("C:/Users/Fred Devereaux/Downloads/Microsoft Windows XP Startup Sound.wav")
                playSound("C:/Users/Fred Devereaux/Downloads/I bet you cant last 2 minutes of a dog whistle.wav")
                return True
        for a in range(9): ## increment the count variable by one each time a turn is taken
            if board[a] == "X" or board[a] == "O":
                count += 1
            if count == 9: ##if the board is full and no winner has been selected, then the game is a tie
                print("The game ends in a Tie\n")
                return True
            
    while not gameEnd: ## while the bool gameEnd is false
        draw() ## draw the board 
        gameEnd = check_board() 
        if gameEnd == True: ##if gameEnd is true after comparing it to the check_board function, break the while loop.
            break
        print("player 1 place an x")
        p1()
        print() ##space it out
        draw()
        gameEnd = check_board()
        if gameEnd == True:
            break
        print("player 2 place a circle")
        p2()
        print()

    if input("play again? (y/n)\n") == "y": ##if the player wants to play again, space then call repeat the bic_bac_boe method
        print()
        bic_bac_boe()

bic_bac_boe()
```
