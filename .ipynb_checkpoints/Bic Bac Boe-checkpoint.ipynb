```python
import pyaudio
import wave
```

```python
def tic_tac_toe():
    board = [1, 2, 3, 4, 5, 6, 7, 8, 9] # array initialized for positions on the board
    gameEnd = False # bool initialized to keep the program running while the game is not over
    win_commbinations = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))

    def draw(): #print each cell in the board
        print(board[0], board[1], board[2])
        print(board[3], board[4], board[5])
        print(board[6], board[7], board[8])
        print()

    def p1(): #function for player one
        n = choose_number()
        if board[n] == "X" or board[n] == "O":
            print("\nYou can't go there. Try again")
            p1()
        else:
            board[n] = "X"

    def p2(): #function for player two
        n = choose_number()
        if board[n] == "X" or board[n] == "O":
            print("\nYou can't go there. Try again")
            p2()
        else:
            board[n] = "O"

    def choose_number():
        while True:
            while True:
                a = input()
                try:
                    a  = int(a)
                    a -= 1
                    if a in range(0, 9):
                        return a
                    else:
                        print("\nThat's not on the board. Try again")
                        continue
                except ValueError:
                   print("\nThat's not a number. Try again")
                   continue

    def check_board():
        count = 0
        for a in win_commbinations:
            if board[a[0]] == board[a[1]] == board[a[2]] == "X":
                print("player 1 wins\n")
                print("Congratulations!\n")
                
                #define stream chunk   
                chunk = 1024  

                #open a wav format music  
                f = wave.open(r"C:/Users/Fred Devereaux/Downloads/Microsoft Windows XP Startup Sound.wav","rb")  
                #instantiate PyAudio  
                p = pyaudio.PyAudio()  
                #open stream  
                stream = p.open(format = p.get_format_from_width(f.getsampwidth()),  
                                channels = f.getnchannels(),  
                                rate = f.getframerate(),  
                                output = True)  
                #read data  
                data = f.readframes(chunk)  

                #play stream  
                while data:  
                    stream.write(data)  
                    data = f.readframes(chunk)  

                #stop stream  
                stream.stop_stream()  
                stream.close()  

                #close PyAudio  
                p.terminate()
                #define stream chunk   
                chunk = 1024  

                #open a wav format music  
                f1 = wave.open(r"C:/Users/Fred Devereaux/Downloads/I bet you cant last 2 minutes of a dog whistle.wav","rb")
                f2 = wave.open(r"C:/Users/Fred Devereaux/Downloads/Microsoft Windows XP Startup Sound.wav","rb")
                #instantiate PyAudio  
                p = pyaudio.PyAudio()  
                #open stream  
                stream1 = p.open(format = p.get_format_from_width(f1.getsampwidth()),  
                                channels = f1.getnchannels(),  
                                rate = f1.getframerate(),  
                                output = True)   
                #read data  
                data1 = f1.readframes(chunk) 
                #play stream  
                while data1:  
                    stream1.write(data1)  
                    data1 = f1.readframes(chunk)

                #stop stream  
                stream1.stop_stream()  
                stream1.close()   
                
                #close PyAudio  
                p.terminate()
                return True

            if board[a[0]] == board[a[1]] == board[a[2]] == "O":
                print("player 2 wins\n")
                print("Congratulations!\n")
                return True
        for a in range(9):
            if board[a] == "X" or board[a] == "O":
                count += 1
            if count == 9:
                print("The game ends in a Tie\n")
                return True

    while not gameEnd:
        draw()
        gameEnd = check_board()
        if gameEnd == True:
            break
        print("Player 1 place an x")
        p1()
        print()
        draw()
        gameEnd = check_board()
        if gameEnd == True:
            break
        print("Player 2 place a circle")
        p2()
        print()

    if input("Play again (y/n)\n") == "y":
        print()
        tic_tac_toe()

tic_tac_toe()
```
